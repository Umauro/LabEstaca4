# Regresión Lineal 

a) Se utiliza la libreria seaborn para visualizar la relación entre las variables G1 y G2 del dataset "student-por.csv".


```python
students = pd.read_csv("student-por.csv",sep=";", usecols=["G1","G2"])
sns.set_style("white")
x=np.arange(0,len(students))
y1=students["G1"]
y2=students["G2"]


sns.regplot(x,y2,lowess=True, line_kws={"color":"magenta", "linewidth":5})
plt.title('Notas Primer periodo' + str())
plt.ylabel('Nota')
plt.xlabel("G1")
plt.show()

sns.regplot(x,y1,lowess=True, line_kws={"color":"green", "linewidth":5})
plt.title('Notas segundo periodo' + str())
plt.ylabel('Nota')
plt.xlabel("G2")
plt.show()

```

Se logra observar en el gráfico que en ambos periodos el rendimiento se comportó de forma similar, mostrando una curva representativa similar a un recta cercana al valor 11 en ambos casos. Esto permite suponer que el promedio entre ambos periodos, la cual se representaria como el promedio entre ambas curvas representativas, debería comportarse de la misma forma.

Si comparamos directamente cada periodo se obtiene la siguiente grafica:

```python
sns.regplot(y2,y1,lowess=True, line_kws={"color":"red", "linewidth":5})
plt.title('Comparación de periodos' + str())
plt.show()
```

De la cual se logra observar una curva representativa de comportamiento lineal en gran parte de su intervalo.

b) Se realizan boxplots para cada una de las variables

```python
plt.figure(figsize=(10, 6))
plt.boxplot(students["G1"])
plt.title('Notas primer periodo' + str())
plt.ylabel('Nota')
plt.xlabel("G1")
plt.show()  
plt.figure(figsize=(10, 6))
plt.boxplot(students["G2"])
plt.title('Notas segundo periodo' + str())
plt.ylabel('Nota')
plt.xlabel("G2")
plt.show()  
```

Se logra observar en ambos boxplot que la mediana (Q2) se presenta dentro del rango de los 10 y los 12.5. Además se logra observar que en ambos casos se presentan outliers. Se puede observar para ambos gráficos que los datos se concentran en torno al valor de aprobacion, habiendo una mayor concentración cerca de los valores menores a este.

c) Se realiza la curva de densidad para cada periodo incluyendo el valor de asimetría:

```python
from scipy.stats import gaussian_kde,skew
density1 = gaussian_kde(y1)
xs = np.linspace(0,20,200)
density.covariance_factor = lambda : .25
density._compute_covariance()
plt.plot(xs,density(xs))
plt.title('Curva de densidad para G1' + str())
plt.show()
print( 'La asimetria para G1 es: {}'.format( skew(y1) ))
density2 = gaussian_kde(y2)
xs = np.linspace(0,20,200)
density.covariance_factor = lambda : .25
density._compute_covariance()
plt.plot(xs,density(xs))
plt.title('Curva de densidad para G2' + str())
plt.show()
print( 'La asimetria para G2 es: {}'.format( skew(y2) ))
```

Se logra observar que la asimetría para ambos periodos es negativa, lo cual indica que las notas esta cargadas mas hacia la izquierda de la curva. Se logra observar además que el comportamiento de ambas curvas es similar, presentando una mayor densidad de datos nuevamente entre 10 y 12.5.
